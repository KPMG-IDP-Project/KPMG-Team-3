```python
import pandas as pd
import re
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
metadata = pd.read_csv('../../wetransfer_b_2022-12-09_1647/B/CLA_meta_from_2018.csv')
```

```python
jc_119 = metadata.loc[metadata['jc_number'] == 119]

themes_119 = ''

for themes in jc_119['themes_nl']:
    themes_119 = themes_119 + str(themes)

themes_119_replace = themes_119.replace('\\', '/')
themes_119_list = themes_119_replace.split('/')
```

```python
word_cloud_119_dict = Counter(themes_119_list)

wordcloud_119 = WordCloud(width=800, height=400).generate_from_frequencies(word_cloud_119_dict)

plt.figure(figsize=(20, 10))
plt.imshow(wordcloud_119, interpolation='bilinear')
plt.axis('off')
plt.show()
```
